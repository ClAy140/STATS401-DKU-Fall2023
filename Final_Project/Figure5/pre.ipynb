```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('all_data.csv')

# Convert the 'Time' column to datetime format
df['Time'] = pd.to_datetime(df['Time'])

# Extract the month and year for grouping purposes
df['YearMonth'] = df['Time'].dt.to_period('M')

# Group by YearMonth and 'Pool' and count the rows
monthly_counts = df.groupby(['YearMonth', 'Pool']).size().reset_index(name='Monthly Count')

# Sort and accumulate the counts
monthly_counts['Accumulated Count'] = monthly_counts.groupby('Pool')['Monthly Count'].cumsum()

# Display the accumulated counts for each type of pool for each month
print(monthly_counts[['YearMonth', 'Pool', 'Accumulated Count']])
# Save the result to a CSV file
monthly_counts[['YearMonth', 'Pool', 'Accumulated Count']].to_csv('accumulated_counts.csv', index=False)
```
